# **Hypothesis Testing using T-test between "Annual_income" and "label"**

```python
import pandas as pd
from scipy.stats import ttest_ind

print("T test between Annual income and label")

approved_incomes = df_cleaned[df_cleaned['label'] == 0]['Annual_income']
rejected_incomes = df_cleaned[df_cleaned['label'] == 1]['Annual_income']

t_statistic, p_value = ttest_ind(approved_incomes, rejected_incomes, equal_var=False)

print(f'T-Statistic: {t_statistic}')
print(f'P-Value: {p_value}')

if p_value < 0.05:
    print('The difference in annual incomes between approved and rejected applicants is statistically significant.')
else:
    print('There is no statistically significant difference in annual incomes between approved and rejected applicants.')

print("\n")
print("T test between Work experience and label")

approved_incomes = df_cleaned[df_cleaned['label'] == 0]['Work_Experience']
rejected_incomes = df_cleaned[df_cleaned['label'] == 1]['Work_Experience']

t_statistic, p_value = ttest_ind(approved_incomes, rejected_incomes, equal_var=False)

print(f'T-Statistic: {t_statistic}')
print(f'P-Value: {p_value}')

if p_value < 0.05:
    print('The difference in work experience between approved and rejected applicants is statistically significant.')
else:
    print('There is no statistically significant difference in work_experience between approved and rejected applicants.')

```

**T test results conclude that Work experience feature is more important than Annual income**
